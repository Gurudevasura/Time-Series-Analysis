## Install Dependence

```python
import pandas as pd
from fbprophet import Prophet
```

## Read in Data and Process Dates

```python
df = pd.read_csv('dataset.csv')
```

```python
df['Year']= df['Time Date'].apply(lambda x: str(x)[-4:])
df['Month']= df['Time Date'].apply(lambda x: str(x)[-6:-4])
df['Day']= df['Time Date'].apply(lambda x: str(x)[:-6])
df['ds'] = pd.DatetimeIndex(df['Year']+'-'+df['Month']+'-'+df['Day'])
```

```python
df = df.loc[(df['Product']==2667437) & (df['Store']=='QLD_CW_ST0203')]
df.drop(['Time Date', 'Product', 'Store', 'Year', 'Month', 'Day'], axis=1, inplace=True)
df.columns = ['y', 'ds']
```

```python
df.head(8)
```

## Train Model

```python
g = Prophet(interval_width = 0.95,daily_seasonality=True)
model = g.fit(df)
```

## forcasting

```python
future = g.make_future_dataframe(periods=100,freq='D')
forecast = g.predict(future)
forecast.head()
```

```python
plot5 = g.plot(forecast)
```

```python
plt6 = g.plot_components(forecast)
```
